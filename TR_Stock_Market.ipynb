### ðŸ“ Footnotes: Indicator Explanations

- **RSI**: Relative Strength Index â€” momentum oscillator indicating overbought/oversold.
- **Stoch**: Stochastic Oscillator â€” compares close to recent high/low range.
- **StochRSI_K / D**: Fast/slow lines of StochRSI; look for crossovers near extremes.
- **MACD / MACDAS**: Measures momentum through moving average crossovers.
- **BB_BWidth**: Bollinger Band Width â€” shows volatility expansion/contraction.
- **BB_%B**: Shows price relative to Bollinger Bands.
- **YF%**: Custom momentum/price-change metric (% from yesterday or a fixed base).
- **MFI**: Money Flow Index â€” volume-weighted RSI.
- **ADX**: Average Directional Index â€” trend strength (not direction).
- **CCI**: Commodity Channel Index â€” deviation from mean price.
- **ROC**: Rate of Change â€” measures speed of price movement.
- **CMF**: Chaikin Money Flow â€” money flow volume over time (volume + price).
- **ATR**: Average True Range â€” measures volatility.
- **FWMA**: Forward Weighted MA â€” gives more weight to recent prices.
- **VWAP**: Volume Weighted Average Price â€” intraday fair value tracker.
- **Supertrend**: Trend-following indicator; filters noise.
- **st_way**: Custom smoothed trend direction estimate.
- **kama**: Kaufman Adaptive Moving Average â€” adjusts to market volatility.
- **Ichimoku Indicators** (Tenkan, Kijun, Senkou A/B, TK_Cross, Cloud_Position): multi-component system identifying trend, momentum, and support/resistance.
- **CPrice**: Current price, often used to confirm breakouts.
- **change**: Percent change over selected time window.

```python
import pandas as pd

def calculate_buy_sell_points(stock_data: pd.Series):
    """
    Calculates buy and sell points for a single stock based on technical indicators.

    Args:
        stock_data (pd.Series): A single row of a DataFrame containing stock indicator data.

    Returns:
        tuple: A tuple containing the total buy points and total sell points.
    """
    buy_points = 0
    sell_points = 0
    
    # RSI: Momentum oscillator indicating overbought/oversold.
    # Buy: Oversold condition.
    if stock_data['RSI'] < 30:
        buy_points += 1
    # Sell: Overbought condition.
    elif stock_data['RSI'] > 70:
        sell_points += 1

    # MFI (Money Flow Index): Volume-weighted RSI.
    # Buy: Oversold condition.
    if stock_data['MFI'] < 20:
        buy_points += 1
    # Sell: Overbought condition.
    elif stock_data['MFI'] > 80:
        sell_points += 1
        
    # MACD: Moving average crossover.
    # Buy: MACD line crosses above signal line.
    if stock_data['MACD'] > stock_data['MACD_signal']:
        buy_points += 1
    # Sell: MACD line crosses below signal line.
    elif stock_data['MACD'] < stock_data['MACD_signal']:
        sell_points += 1

    # Stochastic Oscillator: Compares close price to its high/low range.
    # Buy: Bullish crossover in the oversold zone.
    if stock_data['Stoch_K'] > stock_data['Stoch_D'] and stock_data['Stoch_K'] < 20:
        buy_points += 1
    # Sell: Bearish crossover in the overbought zone.
    elif stock_data['Stoch_K'] < stock_data['Stoch_D'] and stock_data['Stoch_K'] > 80:
        sell_points += 1
        
    # Supertrend: A trend-following indicator.
    # Buy: Upward trend direction.
    if stock_data['st_way'] == 1:
        buy_points += 2  # Stronger signal
    # Sell: Downward trend direction.
    elif stock_data['st_way'] == -1:
        sell_points += 2 # Stronger signal

    # Ichimoku Cloud: Multi-component system for trend identification.
    # Buy: Price is above the cloud.
    if stock_data['Cloud_Position'] == 'Above_Cloud':
        buy_points += 1
    # Sell: Price is below the cloud.
    elif stock_data['Cloud_Position'] == 'Below_Cloud':
        sell_points += 1
        
    # Bollinger Bands (%B): Shows where the price is relative to the bands.
    # Buy: Price is below the lower band.
    if stock_data['BB_%B'] < 0.0:
        buy_points += 1
    # Sell: Price is above the upper band.
    elif stock_data['BB_%B'] > 1.0:
        sell_points += 1
        
    # ADX (Average Directional Index) with CPrice and EMA_200 for trend direction.
    # This checks for a strong trend (ADX > 20) and combines it with a moving average crossover for direction.
    if stock_data['ADX'] > 20 and stock_data['CPrice'] > stock_data['EMA_200']:
        buy_points += 1
    elif stock_data['ADX'] > 20 and stock_data['CPrice'] < stock_data['EMA_200']:
        sell_points += 1

    return buy_points, sell_points

# Example usage with a sample data row from the notebook output
# Replace this with your actual dataframe
sample_data_row = pd.Series({
    'RSI': 63.10, 'MFI': 70.14, 'MACD': 11.14, 'MACD_signal': 9.94, 'Stoch_K': 79.27, 'Stoch_D': 79.97,
    'st_way': 1, 'Cloud_Position': 'Above_Cloud', 'BB_%B': 0.75, 'ADX': 22.66, 'CPrice': 529.00,
    'EMA_200': 495.27
})

buy_score, sell_score = calculate_buy_sell_points(sample_data_row)
print(f"Buy Score: {buy_score}")
print(f"Sell Score: {sell_score}")
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import pandas_ta as ta
import numpy as np
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

def create_dual_charts_from_your_data(stock_list, period="6mo"):
    """
    Create dual charts (Ichimoku Cloud + MACD) using your existing data structure
    
    Parameters:
    stock_list: List of stock symbols
    period: Time period for data ('1y', '6mo', '3mo', etc.)
    """
    
    # Calculate grid dimensions
    n_stocks = len(stock_list)
    n_rows = n_stocks
    n_cols = 2  # Always 2 columns (Ichimoku + MACD)
    
    # Create figure with subplots
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 6*n_rows))
    fig.suptitle(f'Stock Analysis Dashboard - Ichimoku Cloud & MACD', fontsize=16, fontweight='bold')
    
    # Handle single stock case
    if n_stocks == 1:
        axes = axes.reshape(1, -1)
    
    # Dictionary to store all historical data (like in your classic_analysis function)
    stock_dict = {}
    
    for idx, symbol in enumerate(stock_list):
        try:
            print(f"Processing {symbol} ({idx+1}/{n_stocks})...")
            
            # Download stock data (using your method)
            ticker = yf.Ticker(symbol)
            historical_data = ticker.history(period=period)
            
            if historical_data.empty:
                print(f"No historical data for {symbol}, skipping...")
                continue
            
            # Calculate indicators using YOUR method from classic_analysis
            historical_data = calculate_your_indicators(historical_data)
            stock_dict[symbol] = historical_data
            
            # Plot Ichimoku Cloud (Left chart)
            ax_ichimoku = axes[idx, 0] if n_stocks > 1 else axes[0, 0]
            plot_ichimoku_cloud_your_data(ax_ichimoku, historical_data, symbol)
            
            # Plot MACD (Right chart)
            ax_macd = axes[idx, 1] if n_stocks > 1 else axes[0, 1]
            plot_macd_your_data(ax_macd, historical_data, symbol)
            
        except Exception as e:
            print(f"Error processing {symbol}: {str(e)}")
            continue
    
    plt.tight_layout()
    plt.subplots_adjust(top=0.95)  # Adjust for main title
    plt.show()
    
    return fig, stock_dict

def calculate_your_indicators(historical_data):
    """Calculate indicators using YOUR exact method from classic_analysis"""
    
    # RSI
    historical_data['RSI'] = ta.rsi(historical_data['Close'], length=14)
    
    # ADX
    adx_data = ta.adx(historical_data['High'], historical_data['Low'], historical_data['Close'], length=14)
    historical_data['ADX'] = adx_data['ADX_14']
    
    # MACD - using YOUR method
    macd = ta.macd(historical_data['Close'], fast=12, slow=26, signal=9)
    historical_data = pd.concat([historical_data, macd], axis=1)
    
    # MACDAS - your custom indicator
    historical_data['MACDAS'] = historical_data['MACD_12_26_9'] - historical_data['MACDs_12_26_9']
    historical_data['MACDAS_Signal'] = historical_data['MACDAS'].ewm(span=9, adjust=False).mean()
    
    # CCI
    historical_data['CCI'] = ta.cci(historical_data['High'], historical_data['Low'], historical_data['Close'], length=20)
    
    # ROC
    historical_data['ROC'] = ta.roc(historical_data['Close'], length=12)
    
    # ATR
    historical_data['ATR'] = ta.atr(historical_data['High'], historical_data['Low'], historical_data['Close'], length=14)
    
    # MFI
    historical_data['MFI'] = ta.mfi(historical_data['High'], historical_data['Low'], historical_data['Close'], historical_data['Volume'], length=14)
    
    # FWMA
    historical_data["FWMA"] = ta.fwma(historical_data['Close'], length=14)
    
    # OBV
    historical_data['OBV'] = ta.obv(historical_data['Close'], historical_data['Volume'])
    
    # CMF
    historical_data['CMF'] = ta.cmf(historical_data['High'], historical_data['Low'], historical_data['Close'], historical_data['Volume'])
    
    # A/D Line
    historical_data['AD'] = ta.ad(historical_data['High'], historical_data['Low'], historical_data['Close'], historical_data['Volume'])
    
    # VWAP
    historical_data["VWAP"] = ta.vwap(historical_data['High'], historical_data['Low'], historical_data['Close'], historical_data['Volume'])
    
    # KAMA
    historical_data["kama"] = ta.kama(historical_data['Close'], length=14)
    
    # Supertrend
    supertrend = ta.supertrend(historical_data['High'], historical_data['Low'], historical_data['Close'], length=10, multiplier=3)
    historical_data['Supertrend'] = supertrend['SUPERT_10_3.0']
    historical_data['Supertrend_direction'] = supertrend['SUPERTd_10_3.0']
    
    # EMA 200
    historical_data['EMA_200'] = ta.ema(historical_data['Close'], length=200)
    
    # Calculate Ichimoku Cloud - YOUR method
    ichimoku = ta.ichimoku(historical_data['High'], historical_data['Low'], historical_data['Close'], 
                          tenkan=9, kijun=26, senkou=52, offset=26, include_chikou=True)
    
    # Handle Ichimoku data structure - YOUR method
    if isinstance(ichimoku, tuple) and len(ichimoku) >= 2:
        ichi_df1, ichi_df2 = ichimoku[0], ichimoku[1]
        historical_data['Tenkan'] = ichi_df1['ITS_9'] if 'ITS_9' in ichi_df1.columns else None
        historical_data['Kijun'] = ichi_df1['IKS_26'] if 'IKS_26' in ichi_df1.columns else None
        historical_data['Senkou_A'] = ichi_df1['ISA_9'] if 'ISA_9' in ichi_df1.columns else None
        historical_data['Senkou_B'] = ichi_df1['ISB_26'] if 'ISB_26' in ichi_df1.columns else None
        historical_data['Chikou'] = ichi_df2['ICS_26'] if 'ICS_26' in ichi_df2.columns else None
    else:
        # Fallback if structure is different
        historical_data['Tenkan'] = None
        historical_data['Kijun'] = None
        historical_data['Senkou_A'] = None
        historical_data['Senkou_B'] = None
        historical_data['Chikou'] = None
    
    # Calculate Stochastic Oscillator
    stoch = ta.stoch(historical_data['High'], historical_data['Low'], historical_data['Close'], k=14, d=3)
    historical_data['Stoch_K'] = stoch['STOCHk_14_3_3']
    historical_data['Stoch_D'] = stoch['STOCHd_14_3_3']
    
    # Calculate Stochastic RSI
    stochrsi = ta.stochrsi(historical_data['Close'], length=14, rsi_length=14, k=3, d=3)
    historical_data['StochRSI_K'] = stochrsi['STOCHRSIk_14_14_3_3']
    historical_data['StochRSI_D'] = stochrsi['STOCHRSId_14_14_3_3']

    # Calculate Bollinger Bands
    bollinger = ta.bbands(historical_data['Close'], length=20, std=2)
    historical_data['BB_Lower'] = bollinger.iloc[:, 0]  # Lower Band
    historical_data['BB_Middle'] = bollinger.iloc[:, 1]  # Middle Band
    historical_data['BB_Upper'] = bollinger.iloc[:, 2]  # Upper Band
    historical_data['BB_BWidth'] = bollinger.iloc[:, 3]  # Bandwidth
    historical_data['BB_%B'] = bollinger.iloc[:, 4]  # %B
    
    return historical_data

def plot_ichimoku_cloud_your_data(ax, data, symbol):
    """Plot Ichimoku Cloud chart using your data"""
    
    # Plot price
    ax.plot(data.index, data['Close'], label='Price', linewidth=2, color='black')
    
    # Check if Ichimoku data exists
    if data['Tenkan'].isna().all():
        ax.text(0.5, 0.5, f'No Ichimoku data available for {symbol}', 
                horizontalalignment='center', verticalalignment='center', 
                transform=ax.transAxes)
        ax.set_title(f'{symbol} - Ichimoku Cloud', fontweight='bold')
        return
    
    # Plot Ichimoku lines
    if not data['Tenkan'].isna().all():
        ax.plot(data.index, data['Tenkan'], label='Tenkan-sen (9)', color='red', linewidth=1)
    if not data['Kijun'].isna().all():
        ax.plot(data.index, data['Kijun'], label='Kijun-sen (26)', color='blue', linewidth=1)
    if not data['Chikou'].isna().all():
        ax.plot(data.index, data['Chikou'], label='Chikou Span', color='green', linewidth=1, alpha=0.7)
    
    # Plot cloud (Kumo)
    if not data['Senkou_A'].isna().all() and not data['Senkou_B'].isna().all():
        senkou_a = data['Senkou_A'].dropna()
        senkou_b = data['Senkou_B'].dropna()
        
        if not senkou_a.empty and not senkou_b.empty:
            # Align indices
            common_index = senkou_a.index.intersection(senkou_b.index)
            if not common_index.empty:
                senkou_a_aligned = senkou_a.loc[common_index]
                senkou_b_aligned = senkou_b.loc[common_index]
                
                # Fill cloud area
                ax.fill_between(common_index, senkou_a_aligned, senkou_b_aligned,
                               where=(senkou_a_aligned >= senkou_b_aligned), 
                               color='lightgreen', alpha=0.3, label='Bullish Cloud')
                ax.fill_between(common_index, senkou_a_aligned, senkou_b_aligned,
                               where=(senkou_a_aligned < senkou_b_aligned), 
                               color='lightcoral', alpha=0.3, label='Bearish Cloud')
                
                # Plot Senkou lines
                ax.plot(senkou_a.index, senkou_a, label='Senkou Span A', color='orange', linewidth=1, alpha=0.8)
                ax.plot(senkou_b.index, senkou_b, label='Senkou Span B', color='purple', linewidth=1, alpha=0.8)
    
    ax.set_title(f'{symbol} - Ichimoku Cloud', fontweight='bold')
    ax.set_ylabel('Price')
    ax.legend(loc='upper left', fontsize=8)
    ax.grid(True, alpha=0.3)
    ax.tick_params(axis='x', rotation=45)

def plot_macd_your_data(ax, data, symbol):
    """Plot MACD chart using your data structure"""
    
    # Check if MACD data exists
    if 'MACD_12_26_9' not in data.columns or data['MACD_12_26_9'].isna().all():
        ax.text(0.5, 0.5, f'No MACD data available for {symbol}', 
                horizontalalignment='center', verticalalignment='center', 
                transform=ax.transAxes)
        ax.set_title(f'{symbol} - MACD', fontweight='bold')
        return
    
    # Remove NaN values for plotting
    macd_data = data[['MACD_12_26_9', 'MACDs_12_26_9', 'MACDh_12_26_9']].dropna()
    
    if macd_data.empty:
        ax.text(0.5, 0.5, f'No MACD data available for {symbol}', 
                horizontalalignment='center', verticalalignment='center', 
                transform=ax.transAxes)
        ax.set_title(f'{symbol} - MACD', fontweight='bold')
        return
    
    # Plot MACD and Signal lines using YOUR column names
    ax.plot(macd_data.index, macd_data['MACD_12_26_9'], label='MACD', color='blue', linewidth=2)
    ax.plot(macd_data.index, macd_data['MACDs_12_26_9'], label='Signal', color='red', linewidth=2)
    
    # Plot histogram using YOUR column names
    histogram_values = macd_data['MACDh_12_26_9']
    colors = ['green' if x >= 0 else 'red' for x in histogram_values]
    ax.bar(macd_data.index, histogram_values, label='Histogram', 
           color=colors, alpha=0.6, width=1)
    
    # Add zero line
    ax.axhline(y=0, color='black', linestyle='-', alpha=0.5)
    
    ax.set_title(f'{symbol} - MACD', fontweight='bold')
    ax.set_ylabel('MACD')
    ax.legend(loc='upper left')
    ax.grid(True, alpha=0.3)
    ax.tick_params(axis='x', rotation=45)

# Example usage with your portfolio
portfolio = ["MGROS.IS","KCHOL.IS","THYAO.IS","KRDMD.IS","CCOLA.IS",
             "BRSAN.IS","ZOREN.IS","TTRAK.IS","AEFES.IS","FROTO.IS","ALCAR.IS"]

# Create charts for your portfolio
if __name__ == "__main__":
    # Create the dual charts using your data structure
    fig, stock_data_dict = create_dual_charts_from_your_data(portfolio, period="6mo")
    
    # You can access the historical data with all indicators for each stock:
    # stock_data_dict['MGROS.IS'] will contain all the calculated indicators
    
    # Save the figure if needed
    # plt.savefig(f'dual_charts_{datetime.now().strftime("%Y%m%d")}.png', dpi=300, bbox_inches='tight')

print("Dual chart system using your data structure is ready!")
print("Usage: create_dual_charts_from_your_data(portfolio, period='6mo')")
print("The function returns both the figure and a dictionary with all historical data and indicators.")
```

```python
# check_data.py

import yfinance as yf
import pandas_ta as ta
import pandas as pd

def run_sanity_check(symbol='AAPL'):
    """
    Fetches the latest OHLC data and calculates the RSI for a given stock symbol.
    This helps compare local calculations with a platform like TradingView.
    """
    print(f"--- Running Data Sanity Check for: {symbol} ---")
    
    try:
        # --- Step 1: Fetch Data from Yahoo Finance ---
        df = yf.download(symbol, period='1y', interval='1d', progress=False)

        if df.empty:
            print(f"No data downloaded for {symbol}. It might be an invalid ticker or there might be a network issue.")
            return

        # --- Step 2: Calculate RSI ---
        # We use talib=True because the TA-Lib library's implementation of RSI is a
        # widely-used standard and often aligns better with major trading platforms.
        # TradingView's default RSI uses an EMA-based smoothing method, which this mimics more closely.
        df.ta.rsi(length=14, append=True, talib=True)

        # --- Step 3: Get the most recent data for comparison ---
        latest_data = df.iloc[-1]

        print(f"\nDate: {latest_data.name.date()}")
        print("--- OHLCV Data (from yfinance) ---")
        print(f"Open:   {latest_data['Open']:.2f}")
        print(f"High:   {latest_data['High']:.2f}")
        print(f"Low:    {latest_data['Low']:.2f}")
        print(f"Close:  {latest_data['Close']:.2f}")
        print(f"Volume: {latest_data['Volume']:.0f}")

        print("\n--- Indicator Value ---")
        rsi_column_name = 'RSI_14'
        if rsi_column_name in latest_data and pd.notna(latest_data[rsi_column_name]):
            print(f"RSI (14): {latest_data[rsi_column_name]:.2f}")
        else:
            print("RSI could not be calculated. Not enough data points available.")

    except Exception as e:
        print(f"\nAn error occurred. Please ensure 'yfinance' and 'pandas_ta' are installed in your environment.")
        print(f"You can install them using: pip install yfinance pandas_ta")
        print(f"Error details: {e}")

if __name__ == '__main__':
    # You can change the stock symbol here to check any stock you want
    run_sanity_check('AAPL')
```
