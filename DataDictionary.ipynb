## Key Reversion Concepts

### **Mean Reversion vs Trend Following**
- **Best Markets**: Ranging, high-volume, established markets
- **Avoid During**: Strong breakouts, news events, low liquidity periods
- **Time Frames**: Often more effective on shorter timeframes (intraday to weekly)

### **Signal Strength Hierarchy**
1. **Multiple confirmations**: When several indicators align
2. **Volume confirmation**: VWAP + volume delta agreeing
3. **Statistical extremes**: 2+ standard deviations from mean
4. **Pattern + oscillator**: Engulfing pattern with RSI extreme

### **Risk Management for Mean Reversion**
- **Stop losses**: Tight stops since expecting quick reversals
- **Position sizing**: Often smaller positions due to lower win rates
- **Time limits**: Exit if reversion doesn't occur within expected timeframe
