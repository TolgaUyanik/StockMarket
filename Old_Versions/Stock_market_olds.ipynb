```python
# Create a scoring system for finding oversold stocks with recovery potential
def find_recovery_candidates(df, bist100_only=True):
    stocks = df.copy()
    
    if bist100_only:
        stocks = stocks[stocks["symbol"].isin(bist100)]
    
    # Calculate scores for each metric
    stocks['RSI_Score'] = (30 - stocks['RSI']) / 30  # Higher score when RSI is lower (more oversold)
    stocks['BB_Score'] = (0.5 - stocks['BB_%B'])  # Higher score when price is below BB middle
    stocks['ROC_Score'] = (stocks['ROC'] + 10) / 20  # Normalize ROC around -10 to +10 range
    
    # Combined score (weighted)
    stocks['Recovery_Score'] = (
        stocks['RSI_Score'] * 0.4 +    # RSI has highest weight
        stocks['BB_Score'] * 0.3 +     # BB position
        stocks['ROC_Score'] * 0.3      # Recent momentum
    )
    
    # Filter conditions
    recovery_candidates = stocks[
        (stocks['RSI'] < 40) &          # Oversold but not extremely
        (stocks['BB_%B'] < 0.5) &       # Below BB middle line
        (stocks['ADX'] > 15)            # Showing some trend strength
    ]
    
    # Sort by recovery potential score
    recovery_candidates = recovery_candidates.sort_values('Recovery_Score', ascending=False)
    
    # Select relevant columns for display
    columns = ['symbol', 'CPrice', 'RSI', 'BB_%B', 'ROC', 'ADX', 'Recovery_Score']
    
    return recovery_candidates[columns].head(15)

# Use the function
recovery_stocks = find_recovery_candidates(df2)
print("\nTop Recovery Candidates for Next Month:")
display(recovery_stocks)

# Optional: Plot these candidates
plot_portfolio_with_percentage(recovery_stocks['symbol'].tolist(), benchmark="XU100.IS")
```
